```python
import os
from rembg import remove, new_session
from PIL import Image
```

```python
def detoure_abeille_rapide(image_path, output_path):
    print(f"⏳ Traitement en cours : {image_path}")
    
    # 1. Charger l'image originale
    image_originale = Image.open(image_path).convert("RGB")
    
    # 2. Initialiser le modèle léger "u2netp" (télécharge ~4 Mo la première fois)
    session_rapide = new_session("u2netp") 
    
    # 3. Détourer l'image (renvoie une image RGBA transparente)
    image_detouree = remove(image_originale, session=session_rapide)
    
    # 4. Créer un fond noir
    fond_noir = Image.new("RGBA", image_detouree.size, (0, 0, 0, 255))
    
    # 5. Coller l'abeille détourée sur le fond noir
    fond_noir.paste(image_detouree, (0, 0), image_detouree)
    
    # 6. Reconvertir en RGB classique et sauvegarder
    fond_noir.convert("RGB").save(output_path)
    print(f"✅ Fini ! Image sauvegardée ici : {output_path}")

# --- TES CHEMINS DE TEST ---
chemin_test = "/home/oclaich/SDD/Hackaton_abeilles_tigres/data/train/Andrena aliciae/0b2eed58fdc25da54ea0e233e7c893229d443064.jpg" 
chemin_resultat = "/home/oclaich/SDD/Hackaton_abeilles_tigres/background_removal_tests/test_abeille_detouree_rapide.jpg"

# Créer le dossier de destination s'il n'existe pas encore
os.makedirs(os.path.dirname(chemin_resultat), exist_ok=True)

if os.path.exists(chemin_test):
    detoure_abeille_rapide(chemin_test, chemin_resultat)
else:
    print(f"❌ L'image source est introuvable : {chemin_test}")
```
